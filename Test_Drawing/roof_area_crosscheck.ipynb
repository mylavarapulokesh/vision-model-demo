```python
from unstructured.partition.pdf import partition_pdf
from langchain_openai import ChatOpenAI
```

```python
def partition_document(file_path: str):
    """Extract elements from PDF using unstructured"""
    print(f"Partitioning document: {file_path}")
    
    elements = partition_pdf(
        filename=file_path,  # Path to your PDF file
        strategy="hi_res", # Use the most accurate (but slower) processing method of extraction
        infer_table_structure=True, # Keep tables as structured HTML, not jumbled text
        extract_image_block_types=["Image"], # Grab images found in the PDF
        extract_image_block_to_payload=True,
        #chunking_strategy="by_page" # Store images as base64 data you can actually use
    )
    
    print(f"Extracted {len(elements)} elements")
    return elements

# Test with your PDF file
file_path = "Test Drawing.pdf"  # Change this to your PDF path
elements = partition_document(file_path)
```

```python
import os
from typing import Optional
from langchain_openai import ChatOpenAI
from langchain_core.messages import SystemMessage, HumanMessage

def extract_text_from_base64_image(base64_string: str, mime_type: str = "image/jpeg") -> Optional[str]:
    """
    Sends a base64 encoded image to GPT-4o via LangChain to extract all containing text
    while preserving structural layout.
    """
    
    system_prompt = (
        "You are an expert OCR and document analysis engine. Your sole task is to extract "
        "readable text from the provided image accurately and do not extract Diagram Text. Do not extrapolate, do not summarize, "
        "and do not add conversational commentary. Output only the extracted text."
        "concatenate individual words with a space"
    )
    
    user_prompt = (
        "Transcribe all text from this image exactly as it appears. \n"
        "Follow these structural rules:\n"
        "1. Maintain the natural reading order and layout flow (headings, paragraphs, lists).\n"
        "2. If you encounter a table, chart, or matrix, convert it into a clean Markdown table format.\n"
        "3. If text is embedded inside a diagram or infographic, extract it and list it clearly "
        "under a relevant section marker like '[DIAGRAM TEXT]'."
    )

    try:
        # 1. Properly construct the data URL
        image_data_url = f"data:{mime_type};base64,{base64_string}"
        
        # 2. Instantiate the LangChain ChatOpenAI object cleanly
        # Setting temperature=0.0 forces deterministic OCR accuracy
        model = ChatOpenAI(model="gpt-4o", temperature=0.0)
        
        # 3. Use LangChain's native SystemMessage and HumanMessage schemas
        messages = [
            SystemMessage(content=system_prompt),
            HumanMessage(
                content=[
                    {"type": "text", "text": user_prompt},
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": image_data_url,
                            "detail": "high"  # High detail ensures clear resolution for tiny text
                        }
                    }
                ]
            )
        ]
        
        # 4. Invoke the model execution
        response = model.invoke(messages)
        
        # 5. Extract and return just the textual content string from the AIMessage object
        return response.content
        
    except Exception as e:
        print(f"Error during GPT-4o vision execution: {e}")
        return None
```

```python
def textUnderTitle_and_textInImage(elements):

    """ 
    This function returns a list containing 
    1) Text inside the image of Roof Area layout
    2) Text at the bottom right of the page under `title`
    """


    # image_with_roof_area = [] 
    # text_in_image = ''
    text_under_title = ''

    # vision_llm = ChatOpenAI(model="gpt-4o")

    for i,element in enumerate(elements):

        # if 'Image' in str(type(element)) and 'roof area' in element.to_dict().get('text').lower():
        #     text_in_image = element.to_dict().get('text')
        # Check if it's an Image and safely grab the attribute from metadata
        if element.category == 'Image' and hasattr(element.metadata, 'image_base64'):
            image_base64 = element.metadata.image_base64
    
            if image_base64:
                response = extract_text_from_base64_image(image_base64)
        
                if response:
                    # Clean up the marker and strip out any leftover leading/trailing whitespaces
                    cleaned_response = response.replace("[DIAGRAM TEXT]", "").strip()
                    #print(f"content we get from base64 image: {cleaned_response}")

        if 'array layout' in element.to_dict().get('text').lower():
            text_under_title = elements[i+1].to_dict().get('text') + " " + elements[i+2].to_dict().get('text')

    return ["".join(cleaned_response.split()),"".join(text_under_title.split())]

        
```

```python
def seperate_elements_by_page_number(elements):

    """  
    This function returns a list of lists, 
    each list containing the elements with in a single page of pdf
    """

    start_page_number = elements[0].to_dict().get('metadata').get('page_number')
    ending_page_number = elements[-1].to_dict().get('metadata').get('page_number')
    global_list_data = [[] for _ in range(start_page_number,ending_page_number+1)]

    for element in elements:
        page_number = element.to_dict().get('metadata').get('page_number')
        global_list_data[page_number - 1].append(element)
    
    return global_list_data
```

```python
## Implement step by step 

elements_seperated_by_pageno = seperate_elements_by_page_number(elements=elements)

for i,element in enumerate(elements_seperated_by_pageno):
    response = textUnderTitle_and_textInImage(elements=element)
    print(f"roof area details in page: {i+1} is {response}")
    if response[0] == response[1]:
        print(f"entered roof area details are matched in page: {i+1}")
    else:
        print(f"entered roof area details did not match in page: {i+1}")
```
